### Session Duration:
Assuming a session ends when a user logs off or there's no subsequent activity, the session duration would be the sum of time_spent for each session of a user.

```python
# Group by user and calculate total time spent in each session
session_duration = df.groupby('uid')['time_spent'].sum()

# Calculate the average session duration
average_session_duration = session_duration.mean()
print(f"Average Session Duration: {average_session_duration} seconds")
```

## distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the distribution
plt.figure(figsize=(10, 6))
sns.histplot(session_duration, bins=30, kde=True, color="royalblue")

# Adding title and labels
plt.title('Distribution of Session Durations')
plt.xlabel('Session Duration (seconds)')
plt.ylabel('Number of Users')
plt.show()

```

### descriptive stats

MEDIAN: Insight: For session duration, it shows the central point  half of the users spend less time, and half spend more time than this value.

STD DEVIATION: A high standard deviation means values are generally far from the mean; a low standard deviation means they're close to the mean.
Insight: It informs about the variability in session durations. High variability might mean the user experience is inconsistent.

```python
from scipy.stats import kurtosis, skew

mean_duration = session_duration.mean()
median_duration = session_duration.median()
std_deviation = session_duration.std()
excess_kurtosis = kurtosis(session_duration)
skewness_val = skew(session_duration)

print(f"Mean Session Duration: {mean_duration} seconds")
print(f"Median Session Duration: {median_duration} seconds")
print(f"Standard Deviation: {std_deviation} seconds")
print(f"Excess Kurtosis: {excess_kurtosis}")
print(f"Skewness: {skewness_val}")
```

## Frequency of Use:
To measure the frequency of use, we'd need a date or timestamp associated with each row. Assuming time column gives the timestamp when the user visited current_scrn, you can count the distinct days a user was active:

The "Frequency of App Use" section examines how often users engage with the app. This kind of analysis helps to understand user retention, habit formation, and the general appeal of the app to its user base. Some additional analyses in this section could be:

Distribution of User Frequencies: Understanding the distribution can provide insights into how different segments of users engage with the app.

Classification of Users: Based on the frequency, classify users as daily users, weekly users, or monthly users. This allows you to tailor features or marketing campaigns for specific user groups.

Retention Rate Over Time: Track how many users continue to use the app over time, giving insights into the app's long-term appeal.

```python
# Assuming the 'datetime' column is in string format, if not, skip this conversion
df['datetime'] = pd.to_datetime(df['datetime'])

# Extract the date from datetime
df['date'] = df['datetime'].dt.date

# Calculate frequency of use for each user
user_frequency = df.groupby('uid')['date'].nunique()

# For overall frequency (e.g., daily, weekly, monthly averages), you can further analyze user_frequency
average_daily_frequency = user_frequency.mean()
print(f"Average Daily Frequency: {average_daily_frequency}")

'''This part groups the data by the uid (each unique user) and counts
the unique number of dates (nunique()) on which each user interacted with the app. '''

```

Classification of Users:

This classification would ideally require understanding the total duration over which this data was collected. Let's say, for example, the data was collected over a month. Users interacting almost every day can be classified as 'daily', every week as 'weekly', and once or fewer as 'monthly'.

```python
def classify_user(frequency):
    if frequency >= 25:  # Nearly daily in a month
        return 'daily'
    elif frequency >= 4:  # At least once a week in a month
        return 'weekly'
    else:
        return 'monthly'

user_classification = user_frequency.apply(classify_user)
```

```python
import matplotlib.pyplot as plt

# Plotting the histogram
ax = user_classification.value_counts().plot(kind='bar', color='skyblue', figsize=(7,5))

# Adding title and labels
plt.title('User Classification')
plt.xlabel('User Type')
plt.ylabel('Number of Users')

# Annotate the bars with their respective heights (counts)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Retention Rate Over Time:

This metric is typically measured over specific time intervals after a user's first interaction. It tracks the percentage of users who are still active after that interval.

```python
# First interaction date for each user
first_interaction = df.groupby('uid')['date'].min()

# Check if users are still active after, e.g., 7 days
retention_date = first_interaction + pd.Timedelta(days=7)
retained_users = df[df['date'].isin(retention_date)].uid.nunique()

retention_rate = retained_users / df['uid'].nunique()
print(f"7-day Retention Rate: {retention_rate * 100:.2f}%")
```

Churn Rate:

Churn rate can be calculated as the percentage of users who were active in the first period (e.g., the first week) but not in the next (e.g., the second week).

```python
# Get the maximum number of weeks from the dataset
total_weeks = (df['date'].max() - df['date'].min()).days // 7

churn_rates = []
users_who_never_returned = {}

for i in range(total_weeks - 1):  # We subtract 1 because we compare each week with the next
    start_date = df['date'].min() + pd.Timedelta(days=7*i)
    end_date = start_date + pd.Timedelta(days=7)

    current_week_users = df[(df['date'] >= start_date) & (df['date'] < end_date)].uid.unique()
    next_week_users = df[(df['date'] >= end_date) & (df['date'] < end_date + pd.Timedelta(days=7))].uid.unique()

    churned = len(set(current_week_users) - set(next_week_users))
    churn_rate = churned / len(current_week_users) if len(current_week_users) > 0 else 0
    churn_rates.append(churn_rate)

    # Identify users who never logged in again
    future_users = df[df['date'] >= end_date].uid.unique()
    never_returned = set(current_week_users) - set(future_users)
    users_who_never_returned[start_date] = never_returned

# Displaying churn rates
for i, rate in enumerate(churn_rates, 1):
    print(f"Churn Rate from Week {i} to Week {i+1}: {rate * 100:.2f}%")

# Displaying users who never returned after a specific week
for date, users in users_who_never_returned.items():
    print(f"After {date}, {len(users)} users never logged in again.")
```
